```python
df_tracks_silver = spark.read.table("silver_dim_track")
df_albums_silver = spark.read.table("silver_dim_album")
```

```python
from pyspark.sql.functions import col

fact_spotify_songs = (
    df_tracks_silver.alias('t')
    .join(df_albums_silver.alias('al'), 'album_id')
    .select(
        col('t.track_id'),
        col('t.album_id'),
        col('t.artist_name'),
        col('t.track_duration_min'),
        col('al.album_release_date')
    )
)

fact_spotify_songs.write.mode('overwrite').saveAsTable('gold_fact_spotify_songs')
```
